```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'Hiragino Sans'
```

```python
# 初期値と崩壊率を定める
N_0 = 10000
r = 10
```

```python
# FigureとAxesを生成する
fig, ax = plt.subplots(figsize=(5, 3))

# 時間軸を定める
t = np.linspace(0, 1, 100)

# 横軸を時間，縦軸をN(t)としてプロットする
ax.plot(t, N_0 * np.exp(-r * t))
ax.set_xlabel('時刻'); ax.set_ylabel('放射性物質の数')

# Figureを保存する（相対パスを指定）
fig.savefig("./2_decay.png")
```

```python
# 演習1

# 初期値と崩壊率を定める
N_0 = 5.6 * 10**12
r = 1.8 * 10**(-9)
```

```python
# 半減期
T = np.log(2) / r
print(T)
```

```python
print(int(T // 60 // 60 // 24 // 365), "年")
```

```python
import datetime
td = datetime.timedelta(seconds=T)
print(td)
```

```python
# FigureとAxesを生成する
fig, ax = plt.subplots(figsize=(5, 3))

# 時間軸を定める
t = np.linspace(0, 10*int(T), 100)

# 横軸を時間，縦軸をN(t)としてプロットする
ax.plot(t, N_0 * np.exp(-r * t))
ax.set_xlabel('時刻'); ax.set_ylabel('放射性物質の数')

# Figureを保存する（相対パスを指定）
fig.savefig("./2_tritium.png")
```

```python
# 初期温度，環境温度，冷却定数を設定
T_env = 25     # [℃]
T_0 = 90       # [℃]
k = 0.01       # [1/s]
```

```python
# FigureとAxesを生成
fig, ax = plt.subplots(figsize=(5,3))

# 時間軸（0〜60分）
t = np.linspace(0, 1000, 100)
# 温度の時間変化を計算
T = T_env + (T_0 - T_env) * np.exp(-k * t)

# プロット
ax.plot(t, T, label="温度変化")
ax.set_xlabel("時間 [s]")
ax.set_ylabel("温度 [${}^\circ\mathrm{C}$]")
ax.legend()

fig.savefig("./2_cooling.png")
```

```python
# 初期温度，環境温度，冷却定数を設定
T_env = 25     # [℃]
T_0 = 90       # [℃]
k1 = 0.01       # [1/s]
k2 = 0.02       # [1/s]
```

```python
# FigureとAxesを生成
fig, ax = plt.subplots(figsize=(5,3))

# 時間軸（0〜60分）
t = np.linspace(0, 1000, 100)
# 温度の時間変化を計算
T1 = T_env + (T_0 - T_env) * np.exp(-k1 * t)
T2 = T_env + (T_0 - T_env) * np.exp(-k2 * t)

# プロット
ax.plot(t, T1, label=f"温度変化 (k={k1})")
ax.plot(t, T2, label=f"温度変化 (k={k2})")
ax.set_xlabel("時間 [s]")
ax.set_ylabel("温度 [${}^\circ\mathrm{C}$]")
ax.legend()

fig.savefig("./2_cooling2.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 初期条件
T_env = 25   # [℃] 環境温度
T_0 = 90     # [℃] 初期温度

# 比較する冷却定数 [1/s]
k_values = [0.001, 0.002]

# FigureとAxesを生成
fig, ax = plt.subplots(figsize=(5,3))

# 時間軸（0〜1000秒）
t = np.linspace(0, 1000, 200)

# 各kについて温度変化を計算・プロット
for k in k_values:
    T = T_env + (T_0 - T_env) * np.exp(-k * t)
    ax.plot(t, T, label=fr"$k = {k}$")

# 軸ラベル・凡例
ax.set_xlabel("時間 [s]")
ax.set_ylabel(r"温度 [$^\circ\mathrm{C}$]")
ax.legend()
ax.grid(True)

# タイトル
ax.set_title("冷却定数の違いによる温度変化")

# 保存
fig.savefig("./2_cooling_compare.png", dpi=300)
plt.show()
```
