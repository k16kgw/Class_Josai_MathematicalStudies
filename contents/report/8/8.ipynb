```python
import mpmath as mp
q, f = 0.002, 1.1

# f1(x, y) = 0, f2(x, y) = 0 を解きたい
def f1(x, y):
    return q*y - x*y + x*(1 - x)
def f2(x, y):
    return -q*y - x*y + f
# 初期値 (x0, y0) = (0.2, 1.0) からニュートン法で解く
x_star, y_star = mp.findroot((f1, f2), (0, 0))

print("平衡点: x* =", x_star, ", y* =", y_star)
```
